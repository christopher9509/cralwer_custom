### Import library

```python
import requests
from pandas import DataFrame
from bs4 import BeautifulSoup
import re
from datetime import datetime
import os
import pandas as pd
from dateutil.parser import parse
import cloudscraper
```

### 크롤링 가능한 시장 인덱스 리스트
- S&P500, KOSPI200, FTSE100, Dow Jones, Nikkei 225

```python
market_ind = input('크롤링하기 원하는 시장 인덱스를 입력해주세요: ')

if market_ind == 'KOSPI':
    news_url = 'https://www.investing.com/indices/kospi-news'
elif market_ind == 'S&P500':
    news_url = 'https://www.investing.com/indices/us-spx-500-news/'
elif market_ind == 'Dow Jones':
    news_url = 'https://www.investing.com/indices/us-30-news/'
elif market_ind == 'FTSE100':
    news_url = 'https://www.investing.com/indices/uk-100-news/'
elif market_ind == 'Nikkei225':
    news_url = 'https://www.investing.com/indices/jp-ni225-news/'
```

```python
# cloudscraper 초기화 및 news_url에 대해 html을 scraping
scraper = cloudscraper.create_scraper()
html = scraper.get(news_url).content

# block이 우회된 html을 BeautifulSoup으로 parsing
soup = BeautifulSoup(html, 'html.parser')
```

```python
# block 없이 잘 우회하여 가져왔는지 확인해봅니다. 자세한건 아래 링크(in a text editor)를 참고하세요.
soup
```

```python
# 크롤링 결과를 DataFrame으로 저장할 경로를 지정합니다(직접 지정해주세요.)
RESULT_PATH = 'D:/jyp/naver_news_2/'

# news date와 date date에 대해 index(key)를 동일하게 가져가야하기 때문에 2가지로 변수를 나눠서 작성합니다.
idx_1 = 0
idx_2 = 0 

# 초기 페이지 설정 및 크롤링할 페이지 수를 입력받습니다.
cur_page = 1
page_num = int(input('몇 페이지까지 크롤링 할까요? : '))
```

```python
date_text=[]
title_text=[]
url_text = []

while cur_page < page_num:
    
    # date에 대한 정보를 가져옵니다.
    news_date= soup.find_all('span', {'class' : 'articleDetails'})
    date_data = [li.find('span', {'class' : 'date'}) for li in news_date]
    d_data = date_data[:-4]
    
    # title 및 url에 대한 정보를 가져옵니다.
    table = soup.find_all('article', {'class' : 'js-article-item articleItem'})
    a_list = [li.find('div', {'class' : 'textDiv'}) for li in table]
    area_list = [li.find('a', {'class' : 'title'}) for li in a_list]
    area_list = area_list[6:]
    
    # title 및 url에 대한 크롤링 결과를 각각의 list에 저장합니다.
    for i in area_list:
        title_text.append(i.get('title'))
        url_text.append(i.get('href'))
        idx_1 += 1
    
    # 날짜에 대한 크롤링 결과를 각각의 list에 저장합니다.
    for j in d_data:
        date_text.append(j.get_text())
        idx_2 += 1

    # 크롤링 결과를 각각의 dictinary로 저장합니다.
    result= {"date" : date_text , "title" : title_text, "url" : url_text}
    cur_page += 1
    print('현재 크롤링 중인 page number:', cur_page)
    
    # 다음 페이지로 넘어가기 위해 url을 변경합니다.
    news_url_2 = news_url + str(cur_page)
    html = scraper.get(news_url_2).content
    soup = BeautifulSoup(html, 'html.parser')
```

```python
# 크롤링 결과를 DataFrame으로 저장합니다.
df = pd.DataFrame(result)
```

### Preprocessing

```python
# 앞의 '-' 제거
for i in range(len(df['date'])):
    df['date'][i] = df['date'][i][3:]

df = df.drop(labels=range(0, 6), axis=0)
df = df.reset_index(drop=True)

# date parsing
for i in range(len(df['date'])):
    dt = parse(df['date'][i])
    df['date'][i] = dt.date()
```

```python
# 크롤링 및 전처리 결과를 DataFrame으로 저장합니다.
now = datetime.now()
outputFileName = '{}_news_investing.csv'.format(market_ind)
df.to_csv(RESULT_PATH+outputFileName)
```
